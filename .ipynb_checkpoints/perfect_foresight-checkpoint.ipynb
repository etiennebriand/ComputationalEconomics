# Notebook 1: Perfect Foresight Dynamics

We want to solve for the dynamics associated with following pair of non-linear first-order difference equations:

$$
C_t^{-\gamma} = \beta E_t[C_{t+1}^{-\gamma}\left( \alpha K_{t}^{(\alpha-1)}+ (1 -\delta) \right)]
$$

$$
K_t = K_{t-1}^\alpha + (1-\delta)K_{t-1} -C_t
$$

for $K_{-1}$ and $K_T$ given. 

```julia
# Load dependencies
using NLsolve
using LaTeXStrings
using Plots
```

We begin by specifying values for the structural parameters. We then solve for the (non-stochastic) steady-state which gives us the terminal value $K_T$ of our problem.

```julia
# structural parameters
β=.99
δ=0.025
α=0.33
γ=1.;
```

```julia
# (non-stochastic) steady-state
K_ss=(β*α/(1-β*(1-δ)))^(1/(1-α))
I_ss=δ*K_ss
C_ss=K_ss^α-I_ss
λ_ss=C_ss^(-γ);
```

## Algo 1: Shooting Algorithm

We initialize variables shooting algorithm.

```julia
K_init=0.9*K_ss;           #inital capital stock
C_guess=0.9*C_ss;          #initial guess for c_0(*)
C_min=eps()                #minimal feasible consumption
C_max=K_init^α             #maximum feasible consumption
err=1e5;                   #initialize convergence error
tol=1e-5;                  #tolerance criterion for the solution
T=200;                     #number of periods
C_vec=zeros(T);            #consumption path vector
C_vec[1]=C_guess           #guess for c_0(*)
K_vec=zeros(T+1);          #capital path vector
K_vec[1]=K_init;           #intial capital stock
convergence=false          #boolean for convergence
maxit=1000;                #maximum number of iterations
iter=0;                    #initialize iterations counter
```

Run the algorithm.

```julia
# shooting algoritm
while abs(err) > tol && iter <= maxit
    
    iter+=1
    
    C_vec=zeros(T);C_vec[1]=C_guess
    K_vec=zeros(T+1);K_vec[1]=K_init;
    
    try
        for t in 1:T
            # capital transition
            K_vec[t+1] = K_vec[t]^α + (1-δ)*K_vec[t] - C_vec[t]

            # Euler update using K_{t+1}
            if t < T
                mpk_next = α*K_vec[t+1]^(α-1)
                C_vec[t+1] = C_vec[t] * (β*(mpk_next + (1-δ)))^(1/γ)
            end
        end
    catch e
        #deal with numerical errors 
    end
    
    err=(K_vec[end]-K_ss)
    #err=(C_vec[end]-C_ss)    #alternative error measure
    
    if err > 0  #K_T > K_ss, we take C_guess to be our new lower bound on consumption
        C_min=C_guess             #update our lower bound
        C_guess=(C_min+C_max)/2   #update our guess for c_0(*) 
        
    else        #K_T < K_ss, we take C_guess to be our new upper bound on consumption
        C_max=C_guess             #update our lower bound
        C_guess=(C_min+C_max)/2   #update our guess for c_0(*)

    end
    
    if err<tol
        convergence=true
    end;
end;
```

```julia
# information
println("INFO:")
println("Convergence: $convergence.")
println("After $iter iterations.")
println("c_0(*) is $(round(C_vec[1] / C_ss, digits = 5)) of nsss consumption.")
```

```julia
#plot the results
p_capital=plot(1:T,K_vec[1:end-1],grid=false,label=L"\{k_t\}_{t_0}^T",framestyle=:box,legend=:bottomright)
hline!(1:T,[K_ss],label=L"k_{ss}")
title!("Capital")
p_conso=plot(1:T,C_vec,grid=false,label=L"\{c_t\}_{t_0}^T",framestyle=:box,legend=:bottomright)
hline!(1:T,[C_ss],label=L"c_{ss}")
title!("Consumption")
fig = plot(p_capital, p_conso, layout = (1,2), size = (800,400));
display(fig)
```

## Algo 2: Time-path iteration

Even though the **shooting algorithm** only implies guessing a single variable $C_0$, it is not the best way to solve for the transition dynamics under perfect foresight. It turns that solving for the paths $\{C_t\}_{t=0}^T$ and $\{K_t\}_{t=0}^T$ that respect the optimality and terminal conditions with an optimizing routine is superior. We call this method **time-path iteration** (stacked Newton).

```julia
# create a function that stacks the equilibrium conditions and compute their residuals
function stacked_cond(logX,Kinit,T,C_ss,K_ss,γ,α,δ,β)
    
    X=exp.(logX)  #ensure numerical stability
    C=X[1:T-1]
    C_lead=X[2:T]
    K_lag=[K_init;X[T+1:end-2]]
    K=X[T+1:end-1]
    K_end=X[end]
    ERR= [ C.^(-γ)-β.*C_lead.^(-γ).*(α.*K.^(α-1).+(1-δ))
           C_lead[end]-C_ss
           K.-K_lag.^α.-(1-δ).*K_lag.+C
           K_end-K_ss ]
    
    return ERR
end;
```

```julia
# test our function with the shooting algorithm's solution
X=[C_vec;K_vec[2:end]];
err=stacked_cond(log.(X),K_init,T,C_ss,K_ss,γ,α,δ,β);
sum(err)
```

```julia
X0=[repeat([C_ss],T);repeat([K_ss],T)]; #our initial guess is the steady-state
sol = nlsolve(
    X -> stacked_cond(X, K_init, T, C_ss, K_ss, γ, α, δ, β),
    X0;
    method = :trust_region,
    ftol = 1e-8,
    xtol = 1e-8
);
X_sol = sol.zero;
C_sol = exp.(X_sol[1:T]);
K_sol = exp.(X_sol[T+1:end]);
```

```julia
err=stacked_cond(X_sol, K_init, T, C_ss, K_ss, γ, α, δ, β);
sum(err)
```

```julia
# Plot the solution
p_conso=plot(1:T,C_sol,grid=false,label=L"\{c_t\}_{t_0}^T",framestyle=:box,legend=:bottomright)
hline!(1:T,[C_ss],label=L"c_{ss}")
title!("Consumption")
p_capital=plot(1:T,K_sol[1:end],grid=false,label=L"\{k_t\}_{t_0}^T",framestyle=:box,legend=:bottomright)
hline!(1:T,[K_ss],label=L"k_{ss}")
title!("Capital")
fig = plot(p_capital, p_conso, layout = (1,2), size = (800,400));
display(fig) #this is the ''same'' path that we obtained earlier
```
